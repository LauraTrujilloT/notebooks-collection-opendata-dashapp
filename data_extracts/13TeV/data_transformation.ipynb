```python
import pandas as pd
```

```python
data_a = pd.read_csv('hyy_data_sample_a.csv')
data_a = data_a.assign(energy_TeV=13)
data_b = pd.read_csv('hyy_data_sample_b.csv')
data_b = data_b.assign(energy_TeV=13)
data_c = pd.read_csv('hyy_data_sample_c.csv')
data_c = data_c.assign(energy_TeV=13)
```

```python
final_df = pd.concat([data_a, data_b, data_c])
final_df.to_csv('hyy_dataset.csv', index=False)
```
